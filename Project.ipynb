```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Sample DataFrame
# data = df.loc[:, ["TARGET", "CNT_CHILDREN", "AMT_INCOME_TOTAL", "AMT_CREDIT", "AMT_ANNUITY", "AMT_GOODS_PRICE", "CNT_FAM_MEMBERS"]]

# For demonstration, creating random data
np.random.seed(0)
data = pd.DataFrame({
    'TARGET': np.random.randint(0, 2, size=10),
    'CNT_CHILDREN': np.random.randint(0, 5, size=10),
    'AMT_INCOME_TOTAL': np.random.randint(50000, 200000, size=10),
    'AMT_CREDIT': np.random.randint(100000, 500000, size=10),
    'AMT_ANNUITY': np.random.randint(5000, 30000, size=10),
    'AMT_GOODS_PRICE': np.random.randint(50000, 400000, size=10),
    'CNT_FAM_MEMBERS': np.random.randint(1, 6, size=10)
})

# Check for any NaN values
if data.isnull().values.any():
    print("Warning: Data contains NaN values. Filling NaNs with 0 for plotting.")
    data.fillna(0, inplace=True)

# Check data types
print(data.dtypes)

# Calculate the correlation matrix
correlation_matrix = data.corr()

# Create the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar_kws={'label': 'Correlation Coefficient'})

# Set the title and labels
plt.title("Correlation Heatmap", fontsize=14)
plt.xlabel("Features", fontsize=12)
plt.ylabel("Features", fontsize=12)

# Display the plot
plt.show()

```
